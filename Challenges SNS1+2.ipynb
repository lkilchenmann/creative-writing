Step 4: Create a ‘scatterplot’ using the relplot Links to an external site. function, with ‘fare’ as the x-axis and ‘age’ as the y-axis and ‘sex’ as the col.

```python
sns.relplot(titanic, x="fare", y="age", col="sex")
```

Step 5 in a new cell, create the same figure, now add ‘pclass’ as the size.

```python
sns.relplot(titanic, x="fare", y="age", col="sex", size="pclass")
```

Step 6: n a new cell, using the .mappan das method, map the values of the ‘survived’ column to a new column called ‘survived_ind'. Map the ‘0’ value to ‘did not survive’ and the ‘1’ value to ‘survived’.

```python
titanic["survived_ind"] = titanic["survived"].map({0: "did not survive", 1: "survived"})
```

Step 7: Finally, in a new cell recreate the figure, add hue of ‘survived_ind’, then add ‘palette’, using the new column of ‘survived_ind’ with a suitable color for ‘did not survive’ and for ‘survived’ (you can use color names or hex-codes. like this #3A0CA3).

```python
sns.relplot(titanic, x="age", y="fare", hue = "survived_ind", alpha = 0.5, col="survived_ind", palette={"did not survive": "#9ACD32", "survived": "#3A0CA3"}, size="pclass")
```
