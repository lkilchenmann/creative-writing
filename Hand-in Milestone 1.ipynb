I looked up how to extract hashtags and found this: https://openclassrooms.com/en/courses/6532301-introduction-to-natural-language-processing/8082674-bonus-context-extract-information-with-regular-expression & https://stackoverflow.com/questions/50582305/extract-and-count-unique-hashtags-per-row-from-a-pandas-dataframe

```python
import re
```

```python
hashtags = pcos['Title'].str.extractall(r'(#\w+)')[0]
```

```python
hashtags
```

```python
hashtags_df = hashtags.reset_index(drop=True)
```

```python
hashtag_counts = hashtags_df.value_counts().reset_index()
```

```python
hashtag_counts.columns = ['Hashtag', 'Frequency']
```

```python
top_hashtags_df = hashtag_counts.head(10)
print(top_hashtags_df)
```

But now I am running into an issue because some of them aren't related to PCOS and just fertility. I want to make sure I find only PCOS-related tweets.

```python
pcos_related = pcos[pcos['Title'].str.contains(r'\bPCOS\b', case=False, na=False)]

hashtags = pcos_related['Title'].str.extractall(r'(#\w+)')[0]
```

```python
hashtags_df = hashtags.reset_index(drop=True)

hashtag_counts = hashtags_df.value_counts().reset_index()

hashtag_counts.columns = ['Hashtag', 'Frequency']
```

```python
top_hashtags_df = hashtag_counts.head(10)
print(top_hashtags_df)
```

```python
hashtag_counts_noPCOS = hashtag_counts[hashtag_counts['Hashtag'].str.lower() != '#pcos']
top_hashtags_df2 = hashtag_counts_noPCOS.head(10)
print(top_hashtags_df2)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(top_hashtags_df2['Hashtag'], top_hashtags_df2['Frequency'], color='darkblue')
plt.title('Top 10 Hashtags Related to PCOS')
plt.xlabel('Hashtags')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout() 
plt.show()
```
