```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
nltk.download("vader_lexicon")
# Create an instance of SentimentIntensityAnalyzer
model = SentimentIntensityAnalyzer()
```

```python
# Function to get sentiment score
def get_vader_sentiment(text):
    scores = model.polarity_scores(text)
    if scores['compound'] >= 0.05:
        return 'Positive'
    elif scores['compound'] <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
pcos_related['Sentiment'] = pcos_related['Title'].apply(get_vader_sentiment)

pcos_related['YearMonth'] = pd.to_datetime(pcos_related['Date']).dt.to_period('M')

sentiment_monthly = pcos_related.groupby(['YearMonth', 'Sentiment']).size().unstack(fill_value=0)

# Plot
sentiment_monthly.plot(kind='bar', stacked=True, figsize=(12, 6), color=['lightcoral', 'lightgreen', 'lightblue'])
plt.title('Sentiment Analysis Over Time for PCOS Hashtags')
plt.xlabel('Month')
plt.ylabel('Number of Mentions')
plt.xticks(rotation=45)
plt.legend(title='Sentiment', loc='upper left')
plt.grid(axis='y')
plt.show()
```
