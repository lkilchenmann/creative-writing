```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')

# Define a function to clean and preprocess the text
def preprocess_text(text):
    text = text.lower()  # Convert to lower case
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)  # Remove URLs
    text = re.sub(r'\@\w+|\#', '', text)  # Remove mentions and hashtags
    text = re.sub(r'[^a-z\s]', '', text)  # Remove punctuation and numbers
    text = text.strip()  # Remove leading/trailing whitespace
    return text

# Apply the function to your DataFrame's text column
pcos_related['Cleaned_Text'] = pcos_related['Text'].apply(preprocess_text)  # Replace 'Text' with your actual text column name


# Tokenize the text and remove stop words
stop_words = set(stopwords.words('english'))

def tokenize_and_remove_stopwords(text):
    words = text.split()
    return [word for word in words if word not in stop_words]

pcos_related['Tokens'] = pcos_related['Cleaned_Text'].apply(tokenize_and_remove_stopwords)

from sklearn.feature_extraction.text import TfidfVectorizer

# Combine cleaned text for TF-IDF analysis
documents = pcos_related['Cleaned_Text']

# Create the TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=1000)  # Adjust max_features as needed
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

# Get the feature names (words)
tfidf_feature_names = tfidf_vectorizer.get_feature_names_out()
```

```python
# Sum TF-IDF scores for each word
sum_tfidf = tfidf_matrix.sum(axis=0)
words_tfidf = [(tfidf_feature_names[i], sum_tfidf[0, i]) for i in range(len(tfidf_feature_names))]
sorted_words_tfidf = sorted(words_tfidf, key=lambda x: x[1], reverse=True)

# Display the top 10 common keywords
top_keywords = sorted_words_tfidf[:10]
print("Top Keywords: ", top_keywords)
```
